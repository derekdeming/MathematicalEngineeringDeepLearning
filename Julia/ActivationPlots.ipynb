```julia
using Plots, Measures

tanh(u) = (exp(u)-exp(-u))/(exp(u)+exp(-u))
sig(u) = 1/(1+exp(-u))
sig2(u) = 2sig(u)-1
step(u) = u >= 0 ? 1 : -1
relu(u) = u >= 0 ? u : 0
leakyrelu(u) = u >= 0 ? u : ϵ*u
uRange = -4:0.01:4
p1 = plot(uRange,step.(uRange),label="Step", lw =2)
    plot!(uRange,tanh.(uRange),label="Tanh", lw =2)
    plot!(uRange,sig.(uRange),label="Sigmoid", lw =2)
    plot!(uRange,sig2.(uRange),label="Scaled Sigmoid", lw =2)
p2 = plot(uRange,relu.(uRange),label="ReLU", lw =2)
    ϵ = 0.05
    plot!(uRange,leakyrelu.(uRange),label="Leaky ReLU (ϵ=0.05)", lw =2)
    ϵ = 0.1
    plot!(uRange,leakyrelu.(uRange),label="Leaky ReLU (ϵ=0.1)", lw =2)

plot(p1,p2,size=(800,400),legend=:topleft,xlabel="u",ylabel="σ(u)",margin=5mm)
```
