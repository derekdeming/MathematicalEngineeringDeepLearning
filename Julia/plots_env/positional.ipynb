```julia
using Plots
```

```julia
r(i, t) = iseven(i) ? sin(t/M^(2i/p)) : cos(t/M^(2i/p))
```

```julia
M = 10^4
p = 768
T = 4000;
```

```julia
pos_embeddings = [r(i, t) for t in 1:T, i in 1:p];
```

```julia
size(pos_embeddings)
```

```julia
heatmap(pos_embeddings[1:60,1:200]', yflip = false, 
        legend = false, 
        xlabel = "Time index", 
        ylabel = "Embedding vector index",
        c = :thermal,
        guidefontsize = 14)
```
