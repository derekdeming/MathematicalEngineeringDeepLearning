```julia
using Plots, Measures, LaTeXStrings, Distributions
```

```julia

f1(x) = 3(x-0.4)^2 /x^0.1+2-log(x)/5+3
mmm = x_grid[argmin([3(x-0.4)^2 /x^0.1+2-log(x)/5+3 for x in x_grid])]
f2(x) = (x-1.2)^2/x^0.15 + 0.8+3
x_grid=0:0.01:1

p1 = plot(x_grid,f1.(x_grid),label = L"\widetilde{E}_{\textrm{unseen}},~~~\textrm{Expected~generalization~performance}", 
    legend=:outertop, xlabel = "Model Complexity", ylabel="Error",lw=2,ticks=false,c=:red)
plot!(x_grid,f2.(x_grid),label = L"\widetilde{E}_{\textrm{train}},~~~~~\textrm{Expected~training~performance}",lw=2,margin=10mm,c=:green)
vspan!([0,0.25],color = :orange, alpha = 0.2,label = "Under-fitting")
vspan!([0.7,1.0],color = :pink, alpha = 0.2,label = "Over-fitting")
plot!([mmm,mmm],[3.7,6],c=:black,label="Optimal model",linestyle=:dash)
```
